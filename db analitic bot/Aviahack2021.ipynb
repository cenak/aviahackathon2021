```python
import numpy as np
import matplotlib.pyplot as plt
 
rng = np.arange(25)
rnd = np.random.randint(0, 20, size=(3, rng.size))
yrs = 0 + rng
 
fig, ax = plt.subplots(figsize=(13, 7))
ax.stackplot(yrs, rng + rnd, labels=['Устройства воздушного запуска', 'Насосы', 'Авиационные стремянки'])
ax.set_title('Совокупность роста заказа самых частых услуг от времени')
ax.legend(loc='upper left')
ax.set_ylabel('Кол-во заявок')
ax.set_xlim(xmin=yrs[0], xmax=yrs[-1])
fig.tight_layout()
 
plt.plot([0, 30], [60, 60])
plt.show()
```

```python
dpi = 80
fig = plt.figure(dpi = dpi, figsize = (512 / dpi, 384 / dpi) )
mpl.rcParams.update({'font.size': 10})

plt.title('Отношение общего числа заявок к невыполненому')

ax = plt.axes()
ax.yaxis.grid(True, zorder = 1)

xs = range(len(data_names))

plt.bar([x + 0.05 for x in xs], [ d * 0.8 for d in data_values],
        width = 0.2, color = 'red', alpha = 0.7, label = 'Не выполнено',
        zorder = 2)
plt.bar([x + 0.3 for x in xs], data_values,
        width = 0.2, color = 'blue', alpha = 0.7, label = 'Общее число заявок',
        zorder = 2)
plt.xticks(xs, data_names)

fig.autofmt_xdate(rotation = 25)

plt.legend(loc='upper right')
fig.savefig('bars.png')
```
